```python
# 최고기온 변화 그래프
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv') # 97년도 ~ 22년 02월까지 기온
data=csv.reader(f)
next(data) # 맨 윗줄의 데이터 정보를 제외
result=[] # 결과를 담을 리스트생성
print(next(data)) # 맨 윗줄 데이터 정보 출력

for row in data:
    if row[4] != '':
        result.append(float(row[4]))

plt.figure(figsize=(20,5))# (가로,세로)
plt.plot(result,color='b')
plt.show()
f.close()
```

```python
# 97년 ~ 22년 내생일의 최고기온 변화 그래프
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[] #결과를 담을 리스트 생성
y=[] #년도를 담을 리스트 생성

for row in data:
    date=row[0].split('-') # 문자열 나누기 함수 : date 변수는 row[0] 첫번째 줄을 '-' 단위로 나누겠다. (ex)1997-01-20 -> [1997,01,20]
    if date[1] == '01' and date[2] == '20' : # 만약 date 리스트의 2번째 칸이 01 이고 3번째칸이 20이면 (1월 20일)
        y.append(date[0]) #y변수에 date[0]<-- 년도 [1997  <--- 이게 date[0],01,20] 추가
        result.append(float(row[4])) #result 변수에 실수형으로 변환된 최고기온 값을 추가한다.
f.close
plt.figure(figsize=(20,5)) # 인터프린터이기 때문에 먼저 선언을 해준다. 그렇지 않으면 그래프 조절이 안댐
plt.plot(y,result,'r.-',label='max_temp',)
plt.legend(loc=4)
plt.show()
```

```python
#97년~22년도 1월달 최고기온 히스토그램
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

result=[]

for row in data:
    date=row[0].split('-')[1] # date 변수에 row[0] 년도 날짜 데이터를 - 로 분리하고 [1] 두번째 원소값을 저장한다. 두번째 원소값은 월 
    if row[4] != ''and date=='01': #만약 최고기온 칸이 공백이 아니고 date변수의 01 (<--- 1월을 의미) 이라면
        result.append(float(row[4])) #result 리스트에 최고기온을 추가하여라.
plt.figure(figsize=(20,5))# 사이즈 조절
plt.title('01_max_temp')      
plt.hist(result,bins=50,color='g',label='hist_test')
plt.legend()
plt.show
```

```python
#hist 내 생일 평균기온 히스토그램
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

result=[]

for row in data:
    month=row[0].split('-')[1] # 월 정보 저장
    day=row[0].split('-')[2] # 일 정보 저장
    date=month+day  # date 변수에 월과 일 정보 리스트 합치기
    
    if date=='0120': # date에 담은 것이 0120 이라면 (1월20일 이라면)
        result.append(float(row[2])) # result 리스트에 평균기온 데이터를 추가.

plt.title('my birthday')
plt.hist(result,bins=10,color='y',label='0120_temp')
plt.legend()
plt.show()
```

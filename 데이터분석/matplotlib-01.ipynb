```python
import matplotlib.pyplot as plt # matplotlib 라이브러리의 pyplot모듈 사용하고 명칭이 길기 때문에 plt라는 명칭을 as함수를 이용해서 만듬
plt.plot([10,20,30,40])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40],[0,1,2,3])#plt.plot([x축데이터],[y축데이터])
plt.show()#그래프 보여주기
```

```python
# 타이틀 추가
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic') # 라이브러리의 title 함수는 한국어가 안되기 때문에 한글 폰트를 설정한다.
plt.title('테스트') # tilte 제목 넣기
plt.plot([10,20,5,45])
plt.show()
```

```python
#범례 추가
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic') # 라이브러리의 title 함수는 한국어가 안되기 때문에 한글 폰트를 설정한다.
plt.title('테스트') # tilte 제목 넣기
plt.plot([10,20,5,45],label='테스트1')# label: 테스트1이라는 범례를 추가
plt.plot([3,15,6,33],label='테스트2')
plt.legend()# 레이블 값이 범례로 나타나게 하는 함수.괄호안에 loc= 0~10까지 숫자를 입력하면 범례의 위치를 조정할수도 있다.
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color')
plt.plot([1,2,3,4],color='b',label='blue')
plt.plot([0.5,1.5,2.5,3.5],color='y',label='yellow')
#red=r,green=g,blue=b,black=k,yellow=y
plt.legend(loc=10) # 범례 가운데 배치 값 =10
plt.show
```

```python
#그래프 선 모양 및 마커 모양
import matplotlib.pyplot as plt
plt.title('test')
plt.plot([1,2,3,4],'r.:',label='dotted')
plt.plot([5,6,7,8],'g^--',label='dashed')
plt.legend(loc=4)
plt.show()
```

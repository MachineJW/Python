```python
#in연산자 사용하여 경기도 연령별 인구수 그래프 그리기
import csv
import matplotlib.pyplot as plt
f=open('krp.csv')
data=csv.reader(f)
next(data)
x=[] # X축 연령을 담을 리스트 생성
y=[] # y축 인구수를 담을 리스트 생성
n=[] # x축에 연령을 담을때 문자열을 분리하여 숫자로된 데이터를 담을 리스트

for row in data:
    if row[0] in '경기도' and row[1] != '계': # 조건: row[0] 첫번째 줄이 '경기도'라는 문자열을 포함하고 row[1]에서 '계'(합계)가 아니라면
        n=row[1].split('세')[0] #0세~100세까지라고 된 문자열을 '세'로 구분하여 앞 숫자부분만 n에 담는다.
        x.append(int(n)) #x축 리스트에 연령숫자데이터를 정수형으로 담고
        y.append(int(row[2])) #y축 리스트에 인구수를 정수형으로 담는다.
        #여기서 정수나 실수형으로 바꿔주지 않는다면 그래프를 그릴때 문자열로 인식하여 그래프가 매우 지저분해진다.

#리스트에 데이터가 정상적으로 담겼는지 확인하기 위하여 그래프를 그리기전에,x와y를 출력해본다.
print(x)
print(y)

#matplot의 pyplot 도구로 그래프 그리기
plt.figure(figsize=(10,5)) #사이즈 지정
plt.rc('font',family='Malgun Gothic') #한글을 사용하기 위하여 폰트설정을 맑은 고딕체로 수정해준다.
plt.title('경기도 연령별 인구수(2022년 01월 기준)') #타이틀 제목 지정
plt.plot(x,y,'g',label='인구수') # x축,y축을 지정해주고 색을 g (green)으로 지정, 범례는 인구수로 저장해준다.
plt.legend()# 범례를 나타내기위한 함수
plt.show() # 마지막으로 그래프 나타내기.
```

```python
#bar그래프 부산광역시 연령별 남성 인구
import csv
import matplotlib.pyplot as plt
f=open('krp.csv')
data=csv.reader(f)
next(data)
x=[]
y=[]
k=[]

for row in data:
    if row[0] in '부산광역시' and row[1] !='계':
        k=row[1].split('세')[0]
        x.append(int(k))
        y.append(int(row[3]))
        
print(x)
print(y)

plt.figure(figsize=(10,5))
plt.rc('font',family='Malgun Gothic')
plt.title('부산광역시 연령별 인구 수 (22년도 01월 기준)')
plt.bar(x,y,color='pink')
plt.show()
```

```python
#input활용,원하는 지역 연령별 인구 찾기
import csv
import matplotlib.pyplot as plt
f=open('krp.csv')
data=csv.reader(f)

x=[]
y=[]
k=[]
print('찾기 가능한 목록:\n 전국,서울특별시,부산광역시,대구광역시,인천광역시,광주광역시,대전광역시,울산광역시,세종특별자치시,\n 경기도,강원도,충청북도,충청남도,전라북도,전라남도,경상북도,경상남도,제주특별자치도')

name=input('찾을 지역의 이름을 입력하세요(행정구역): ')


for row in data:
    if row[0] in name and row[1] !='계':
        k=row[1].split('세')[0]
        x.append(int(k))
        y.append(int(row[3]))
        
plt.figure(figsize=(10,5))
plt.rc('font',family='Malgun Gothic')
plt.title(name + '연령별 인구 수 (22년도 01월 기준)')
plt.bar(x,y,color='g')
plt.show()
```

```python
#파이차트(남성인구와 여성인구 비교)
import csv
import matplotlib.pyplot as plt
f=open('krp.csv')
data=csv.reader(f)
next(data)

m=[] # 남자 인구수를 담을 데이터
w=[] # 여자 인구수를 담을 데이터 

print('찾기 가능한 목록:\n 전국,서울특별시,부산광역시,대구광역시,인천광역시,광주광역시,대전광역시,울산광역시,세종특별자치시,\n 경기도,강원도,충청북도,충청남도,전라북도,전라남도,경상북도,경상남도,제주특별자치도')
name = input('남성과 여성의 인구 비교를 찾을 지역을 입력하세요: ')

for row in data:
    if row[0] in name and row[1]=='계': # 총 합산된 값을 알아야 하므로 row[1]에서 합산된 리스트인 '계'인 경우로 조건을 두었다.
        m.append(int(row[3])) #row[3] 남성인구수
        w.append(int(row[4])) #row[4] 여성인구수
        size=m+w #size 리스트에 m과 w의 값을 담는다.
print(size)# 결과가 잘 담겼는지 확인
#Pie차트
plt.rc('font',family='Malgun Gothic')
plt.axis('equal') # axis를 이용하여 원을 동그랗게 만들어준다. 이것을 설정하지 않으면 원이 타원형으로 나온다.
plt.title(name+ '남녀 성별 비율 (22년도 01월 기준)')
plt.pie(size,labels=['남','여'],autopct='%.1f%%',colors=['skyblue','hotpink'],startangle=90)
#size는 남자와 여자의 인구수를 담은 값, acutopct함수는 퍼센트를 나타낼때 속성값을 지정한다. '%.1f%%'는  소수점 아래 첫번째 소수점까지 표현하겠다는 의미.
#startangle은 그려질 각이 최초로 시작되는 지점을 나타낸다. 90값으로 맞춰주면 12시 정각위치에서 파이차트 그래프가 그려진다.
plt.show()
```
